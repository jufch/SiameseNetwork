## Rename the files

The files will be added the real name of their category. It will be easier to understand the data.

```python
def modify_prefix_image_names(base_path, categories_with_prefixes):
    for category, prefix in categories_with_prefixes.items():
        # Define the full path to the category directory
        category_path = os.path.join(base_path, category.replace(prefix, prefix.replace(" ", "")))
        if os.path.isdir(category_path):
            for image_name in os.listdir(category_path):
                # Define the new image name with the correct prefix
                new_image_name = f"{prefix}_{image_name}" if not image_name.startswith(f"{prefix}_") else image_name
                # Define the current and new full paths for the image
                current_image_path = os.path.join(category_path, image_name)
                new_image_path = os.path.join(category_path, new_image_name)
                # Rename the image
                os.rename(current_image_path, new_image_path)
                print(f"Renamed {image_name} to {new_image_name}")

# Define the base path to the categories directory
base_path = os.path.join(os.path.dirname(os.getcwd()), "OpenSARShip", "Categories")  # Update this path as necessary

# Specify the categories and their desired prefixes
categories_with_prefixes = {
    "Cargo": "Cargo",
    "Bulk Carrier": "BulkCarrier",
    "Container Ship": "ContainerShip"
}

# Call the function
modify_prefix_image_names(base_path, categories_with_prefixes)

```

## Split the data

```python
def split_data(base_path, categories, train_size=0.8, validation_size=0.1):
    # Assuming test_size is the remainder
    test_size = 1 - train_size - validation_size
    
    all_files = []

    # Gather all file paths
    for category in categories:
        category_path = os.path.join(base_path, category)
        files = [os.path.join(category, file) for file in os.listdir(category_path)]
        all_files.extend(files)
    
    # Split the data
    train_files, test_files = train_test_split(all_files, train_size=train_size + validation_size, test_size=test_size, random_state=42)
    train_files, validation_files = train_test_split(train_files, train_size=(train_size / (train_size + validation_size)), random_state=42)
    
    # Create CSV files for each set
    create_csv('train.csv', train_files)
    create_csv('validation.csv', validation_files)
    create_csv('test.csv', test_files)

def create_csv(filename, file_paths):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['file_path'])
        for path in file_paths:
            writer.writerow([path])

# Define the base path to your categories directory and the categories you want to include in the split
base_path = os.path.join(os.path.dirname(os.getcwd()), "OpenSARShip", "Categories")    # Update this path as necessary
categories = ['Cargo', 'Bulk Carrier', 'Container Ship']  # Update this list as necessary

# Perform the split
split_data(base_path, categories)
```

## VV-polarization

```python
def split_data_vv_only(base_path, categories, train_size=0.8, validation_size=0.1, seed=42):
    train_files, validation_files, test_files = [], [], []
    
    # Perform split for each category
    for category in categories:
        category_files = [os.path.join(category, file) for file in os.listdir(os.path.join(base_path, category)) if file.endswith('_vv.tif')]
        train_files_cat, test_val_files_cat = train_test_split(category_files, train_size=train_size, random_state=seed)
        validation_files_cat, test_files_cat = train_test_split(test_val_files_cat, train_size=validation_size/(1-train_size), random_state=seed)
        
        train_files.extend(train_files_cat)
        validation_files.extend(validation_files_cat)
        test_files.extend(test_files_cat)
    
    # Debugging prints
    print(f"Total files: {len(train_files) + len(validation_files) + len(test_files)}")
    print(f"Train files: {len(train_files)}")
    print(f"Validation files: {len(validation_files)}")
    print(f"Test files: {len(test_files)}")
    
    # Create CSV files for each set
    create_csv('train.csv', train_files)
    create_csv('validation.csv', validation_files)
    create_csv('test.csv', test_files)

def create_csv(filename, file_paths):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['file_path'])
        for path in file_paths:
            writer.writerow([path])

# Define the base path to your categories directory and the categories you want to include in the split
base_path = os.path.join(os.path.dirname(os.getcwd()), "OpenSARShip", "Categories")  # Update this path as necessary
categories = ['Cargo', 'Bulk Carrier', 'Container Ship']  # Adjust category names as needed

# Perform the split
split_data_vv_only(base_path, categories)
```

## Cration of frugal data

Dans cette partie, nous allon créer un jeu de données frugal. C'est à dire que nous allons réduire le nombre de données pour chaque catégorie.

```python
import os
import csv
from sklearn.model_selection import train_test_split

def create_frugal_dataset(base_path, categories, sizes, train_size=0.6, validation_size=0.2):
    """
    Create a frugal dataset with specified maximum sizes for each category and split into train, validation, and test sets.

    Args:
    - base_path (str): Base directory path where categories are stored.
    - categories (list): List of category names.
    - sizes (dict): Maximum sizes for each category.
    - train_size (float): Proportion of the dataset to include in the train split.
    - validation_size (float): Proportion of the dataset to include in the validation split.
    """
    all_files = []

    # Gather and limit file paths per category
    for category in categories:
        category_path = os.path.join(base_path, category)
        files = [os.path.join(category, file) for file in os.listdir(category_path)]
        limited_files = files[:sizes[category]]  # Limit the number of files per category
        all_files.extend(limited_files)

    # Calculate test size based on the remaining portion
    test_size = 1 - train_size - validation_size

    # Split the data
    train_files, test_files = train_test_split(all_files, train_size=train_size + validation_size, test_size=test_size, random_state=42)
    train_files, validation_files = train_test_split(train_files, train_size=(train_size / (train_size + validation_size)), random_state=42)

    # Create CSV files for each set
    create_csv('train.csv', train_files)
    create_csv('validation.csv', validation_files)
    create_csv('test.csv', test_files)

def create_csv(filename, file_paths):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['file_path'])
        for path in file_paths:
            writer.writerow([path])

# Define the base path to your categories directory and the categories you want to include in the split
base_path = os.path.join(os.path.dirname(os.getcwd()), "OpenSARShip", "Categories") # Update this path as necessary
categories = ['Cargo', 'Bulk Carrier', 'Container Ship']  # Update this list as necessary
sizes = {'Cargo': 600, 'Bulk Carrier': 250, 'Container Ship': 200}  # Define the max sizes

# Perform the split with frugal dataset constraints
create_frugal_dataset(base_path, categories, sizes)
```
